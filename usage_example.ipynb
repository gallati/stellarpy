```python
from stellarpy import Star, error_table, find_minimum
```

# StellarPy: Usage Example Jupyter Notebook

Initializing a Star object

```python
star = Star(Mtot=5.0, Rtot=11.5, Ltot=40.0, Tc=1.5, X=0.80, Y=0.16)
star
```

Printing total relative error of the calculation.

```python
print(star.error())
```

Finding the minimum relative error using error_table.

```python
error_table(star=star, n=5, dR=0.5, dL=5.0, numbering=True)
```

Finding the minimum relative error using find_minimum and redefining star atributes.

```python
Rmin, Lmin, Tmin, error_min = find_minimum(star=star, x0=None)
star.redefine(Rtot=Rmin, Ltot=Lmin, Tc=Tmin)
```

Now that the star is well defined, let's plot some cool graphs! First, variables throughout the star

```python
star.visualize(merge=True, figsize=(10, 6))
```

Plotting both energy generation rate and opacity

```python
star.visualize(which=["kappa", "epsilon"], normalize=False)
```

Plotting the Teperature-Density Diagram

```python
star.TDD()
```

Plotting Altair in the Hertzsprung–Russell Diagram

```python
star.HR()
```

Accessing star data

```python
star.get()
```
